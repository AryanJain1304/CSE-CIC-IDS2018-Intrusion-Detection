```python
lstm = np.load('/kaggle/input/cse-cic-ids2018-extracted-features/lstm_features.npy', allow_pickle=True).item()
cnn = np.load('/kaggle/input/cse-cic-ids2018-extracted-features/cnn_features.npy', allow_pickle=True).item()
sae = np.load('/kaggle/input/cse-cic-ids2018-extracted-features/sae_features.npy', allow_pickle=True).item()
```

```python
X_train_combined = np.concatenate([lstm['train'], cnn['train'], sae['train']], axis=1)
X_test_combined = np.concatenate([lstm['test'], cnn['test'], sae['test']], axis=1)
```

```python
np.save('X_train_combined.npy', X_train_combined)
np.save('X_test_combined.npy', X_test_combined)
```
